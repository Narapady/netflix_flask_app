```python
import pandas as pd
```

```python
df = pd.read_csv("titles.csv")
df.head()
```

```python
df.shape
```

```python
len(set(df["id"].to_list()))
```

```python
df.info()
```

```python
df["id"] = df["id"].apply(lambda x: x[2:])
df.head()
```

```python
df["genres"] = df["genres"].apply(
    lambda x: x.replace("[", "").replace("]", "").replace("'", "")
)
df.head()
```

```python
df["production_countries"] = df["production_countries"].apply(
    lambda x: x.replace("[", "").replace("]", "").replace("'", "")
)
df.head()
```

```python
df = df.drop(columns=["imdb_id", "tmdb_popularity", "tmdb_score"])
df.head()
```

```python
df.to_csv("titles_cleansed.csv", index=False)
```

```python
df = pd.read_csv("credits.csv")
df.head()
```

```python
df["title_id"] = df["title_id"].apply(lambda x: x[2:])
df.head()
```
