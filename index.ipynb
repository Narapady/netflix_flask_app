```python
import pandas as pd
```

```python
df = pd.read_csv("./data/titles.csv")
df.head()
```

```python
df["id"] = df["id"].str.replace("ts", "").str.replace("tm", "").astype("int")
df.head()
```

```python
df["genres"] = (
    df["genres"]
    .str.replace("[", "")
    .str.replace("]", "")
    .str.replace("'", "")
    .astype("str")
)
df.head()
```

```python
df["production_countries"] = (
    df["production_countries"]
    .str.replace("[", "")
    .str.replace("]", "")
    .str.replace("'", "")
    .astype("str")
)
df.head()
```

```python
df["seasons"] = df["seasons"].fillna(0).astype(int)
df.head()
```

```python
df_title = df.drop(columns=["imdb_id", "tmdb_popularity", "tmdb_score"])
df_title.head()
```

```python
df_credit["title_id"] = (
    df_credit["title_id"].str.replace("ts", "").str.replace("tm", "")
)
df_credit.head()
```
